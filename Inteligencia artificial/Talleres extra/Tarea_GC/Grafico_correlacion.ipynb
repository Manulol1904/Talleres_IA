## Grafico de correlaci√≥n


### Creado a partir de un Dataframe creado por codigo


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creamos un conjunto de datos de ejemplo
df = pd.DataFrame({
    "edad": [18, 22, 25, 30, 35, 40, 50],
    "ingresos": [800, 1200, 2000, 3000, 4000, 5000, 7000],
    "gastos": [500, 700, 1200, 1800, 2500, 3200, 4000]
})

# ===== MATRIZ DE CORRELACI√ìN =====
correlacion = df.corr()
print(correlacion)  # calcula la correlaci√≥n entre todas las columnas
sns.heatmap(correlacion, annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de correlaci√≥n")
plt.show()

# ===== DIAGRAMA DE DISPERSI√ìN =====
sns.scatterplot(data=df, x="edad", y="ingresos")
plt.title("Relaci√≥n entre Edad e Ingresos")
plt.show()
```
